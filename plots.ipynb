```python
# load dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

red_wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep= ';')
white_wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv', sep= ';')
```

```python
red_wine['quality_label'] = red_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')
```

```python
white_wine['wine_type'] = 'white'
white_wine['quality_label'] = white_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

white_wine['quality_label'] = pd.Categorical(white_wine['quality_label'],
categories=['low', 'medium', 'high'])
```

```python
# combine two dataframe (red_wine) & (white_wine)
wines = pd.concat([red_wine, white_wine])

```

```python
# re-shuffle records just to randomize data points
wines = wines.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
wines['wine_type'] = wines['wine_type'].replace(np.nan, 'red')
```

```python
# Red wine plots
# Let's do some plotting to know how the data columns are distributed in the dataset

#Here we see that fixed acidity does not give any specification to classify the quality.
fig = plt.figure(figsize = (10,6))
#sns.barplot(x = 'quality', y = 'fixed acidity', data = red_wine);
#sns.violinplot(x = 'quality', y = 'fixed acidity', data = red_wine)
 sns.swarmplot(x = 'quality', y = 'fixed acidity', data = red_wine,color = 'white');
#Here we see that fixed acidity does not give any specification to classify the quality.
```

```python
#Here we see that its quite a downing trend in the volatile acidity as we go higher the quality 
fig = plt.figure(figsize = (10,6))
#sns.barplot(x = 'quality', y = 'volatile acidity', data = red_wine);
sns.jointplot(x="quality", y="volatile acidity", data=red_wine, kind = 'reg');
```

```python
#Composition of citric acid go higher as we go higher in the quality of the wine
fig = plt.figure(figsize = (10,6))
#sns.barplot(x = 'quality', y = 'citric acid', data = red_wine);
sns.jointplot(x="quality", y="citric acid", data=red_wine, kind = 'reg');
```

```python
fig = plt.figure(figsize = (10,6))
# sns.barplot(x = 'quality', y = 'residual sugar', data = red_wine);
# here residual sugar does not give much impact to the quality of the wine
sns.jointplot(x="quality", y="residual sugar", data=red_wine, kind = 'reg');
```

```python
#Composition of chloride also go down as we go higher in the quality of the wine
fig = plt.figure(figsize = (10,6))
# sns.barplot(x = 'quality', y = 'chlorides', data = red_wine);
sns.jointplot(x="quality", y="chlorides", data=red_wine, kind = 'reg');
```

```python
# here free sulfur dioxide is high on the medium level of the quality of the wine
fig = plt.figure(figsize = (10,6))
#sns.barplot(x = 'quality', y = 'free sulfur dioxide', data = red_wine);
#sns.boxplot(x="quality", y="free sulfur dioxide", data=red_wine, hue = 'quality_label');
sns.jointplot(x="quality", y="free sulfur dioxide", data=red_wine, kind = 'reg');
```

```python
# here total sulfur dioxide is high on the medium level of the quality of the wine
fig = plt.figure(figsize = (10,6))
#sns.barplot(x = 'quality', y = 'total sulfur dioxide', data = red_wine);
# sns.boxplot(x="quality", y="total sulfur dioxide", data=red_wine, hue = 'quality_label');
sns.jointplot(x="quality", y="total sulfur dioxide", data=red_wine, kind = 'reg');
```

```python
#Sulphates level goes higher with the quality of wine
fig = plt.figure(figsize = (10,6))
# sns.barplot(x = 'quality', y = 'sulphates', data = red_wine);
sns.jointplot(x="quality", y="sulphates", data=red_wine, kind = 'reg');
```

```python
#Alcohol level also goes higher as te quality of wine increases
fig = plt.figure(figsize = (10,6))
#sns.barplot(x = 'quality', y = 'alcohol', data = red_wine);
#sns.boxplot(x="quality", y="alcohol", data=red_wine, hue = 'quality_label');
sns.jointplot(x="quality", y="alcohol", data=red_wine, kind = 'reg');
```

```python
# here we see the pH is slightly decreasing according to the quality of the wine
fig = plt.figure(figsize = (10,6))
# sns.barplot(x = 'quality', y = 'sulphates', data = red_wine);
# sns.jointplot(x="quality", y="pH", data=red_wine, kind = 'reg');
sns.swarmplot(x="quality", y="pH", data=red_wine);
```
